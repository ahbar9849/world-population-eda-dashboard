```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/world_population.csv")
df = df.dropna()
df['Population'] = df['Population'].astype(int)
df['Area (km²)'] = df['Area (km²)'].replace(0, pd.NA).fillna(df['Area (km²)'].mean())
df['Density'] = df['Population'] / df['Area (km²)']

top10 = df.nlargest(10, 'Population')
bottom10 = df.nsmallest(10, 'Population')
top_density = df.nlargest(10, 'Density')
bottom_density = df.nsmallest(10, 'Density')
growth_sorted = df.sort_values('Growth Rate', ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(data=top10, x='Population', y='Country', palette="viridis")
plt.title("Top 10 Most Populated Countries")
plt.show()

plt.figure(figsize=(12,6))
sns.barplot(data=bottom10, x='Population', y='Country', palette="magma")
plt.title("Top 10 Least Populated Countries")
plt.show()

plt.figure(figsize=(12,6))
sns.barplot(data=top_density, x='Density', y='Country', palette="coolwarm")
plt.title("Top 10 Most Densely Populated Countries")
plt.show()

plt.figure(figsize=(12,6))
sns.barplot(data=bottom_density, x='Density', y='Country', palette="Blues_r")
plt.title("Top 10 Least Densely Populated Countries")
plt.show()

plt.figure(figsize=(12,6))
sns.barplot(data=growth_sorted.head(10), x='Growth Rate', y='Country', palette="crest")
plt.title("Top 10 Countries by Growth Rate")
plt.show()

plt.figure(figsize=(12,6))
sns.barplot(data=growth_sorted.tail(10), x='Growth Rate', y='Country', palette="flare")
plt.title("Bottom 10 Countries by Growth Rate")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df['Population'], bins=20, kde=True, color="purple")
plt.title("Population Distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(df['Density'], bins=20, kde=True, color="orange")
plt.title("Population Density Distribution")
plt.show()

if 'Growth Rate' in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df['Growth Rate'], bins=20, kde=True, color="skyblue")
    plt.title("Growth Rate Distribution")
    plt.show()

if 'Rank' in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df['Rank'], bins=20, color="green")
    plt.title("Country Rank Distribution")
    plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Population Features")
plt.show()

plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x="Area (km²)", y="Population", size="Growth Rate", hue="Density", sizes=(50,500), palette="viridis", alpha=0.7)
plt.title("Population vs Area with Growth Rate and Density")
plt.show()

plt.figure(figsize=(12,6))
sns.lineplot(data=df.sort_values("Rank"), x="Rank", y="Population", marker="o", color="blue")
plt.title("Population by Rank")
plt.show()

plt.figure(figsize=(12,6))
sns.lineplot(data=df.sort_values("Rank"), x="Rank", y="Density", marker="o", color="red")
plt.title("Density by Rank")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/world_population.csv")
df = df.dropna()
df['Population'] = df['Population'].astype(int)
df['Area (km²)'] = df['Area (km²)'].replace(0, pd.NA).fillna(df['Area (km²)'].mean())
df['Density'] = df['Population'] / df['Area (km²)']

top10 = df.nlargest(10, 'Population')
bottom10 = df.nsmallest(10, 'Population')
top_density = df.nlargest(10, 'Density')
bottom_density = df.nsmallest(10, 'Density')
growth_sorted = df.sort_values('Growth Rate', ascending=False)

sns.set_theme(style="whitegrid")
plt.rcParams.update({'axes.titlesize':12, 'axes.labelsize':10, 'xtick.labelsize':8, 'ytick.labelsize':8})

fig, axes = plt.subplots(3, 3, figsize=(22, 18))
axes = axes.flatten()

sns.barplot(data=top10, x='Population', y='Country', palette="viridis", ax=axes[0])
axes[0].set_title("Top 10 Most Populated")

sns.barplot(data=bottom10, x='Population', y='Country', palette="rocket", ax=axes[1])
axes[1].set_title("Bottom 10 Least Populated")

sns.barplot(data=top_density, x='Density', y='Country', palette="crest", ax=axes[2])
axes[2].set_title("Top 10 Most Dense")

sns.barplot(data=bottom_density, x='Density', y='Country', palette="flare", ax=axes[3])
axes[3].set_title("Bottom 10 Least Dense")

sns.histplot(df['Growth Rate'], bins=20, kde=True, color="skyblue", ax=axes[4])
axes[4].set_title("Growth Rate Distribution")

sns.histplot(df['Population'], bins=20, kde=True, color="purple", ax=axes[5])
axes[5].set_title("Population Distribution")

sns.heatmap(df.corr(numeric_only=True), annot=False, cmap="coolwarm", ax=axes[6], cbar=True)
axes[6].set_title("Correlation Heatmap")

sns.scatterplot(data=df, x="Area (km²)", y="Population",
                size="Growth Rate", hue="Density",
                sizes=(20,200), palette="viridis", alpha=0.6, ax=axes[7])
axes[7].set_title("Population vs Area")
axes[7].set_xscale("log")
axes[7].set_yscale("log")

sns.lineplot(data=df.sort_values("Rank"), x="Rank", y="Population", color="blue", ax=axes[8])
axes[8].set_title("Population by Rank")

plt.tight_layout()
plt.show()

```
